## __PyTessearct__

Python-tesseract is an optical character recognition (OCR) tool for python. That is, it will recognize and “read” the text embedded in images.

Note: pytesseract does not provide true Python bindings. Rather, it simply provides an interface to the tesseract binary

```python
import cv2
import pytesseract
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
def ocr_core(img):
    text = pytesseract.image_to_string(img)
    return text
```

```python
def pre_processing(image):
    # conveert the iamge to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # thresholding the image
    image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
    # noise removal
    image = cv2.medianBlur(image, 5)
    return image
```

```python
# read and displaay the image
img = cv2.imread('example_01.png')
imshow(img)
```

```python
# pre-process and displaay the image
pp_img = pre_processing(img)
imshow(pp_img)
```

```python
# run the occ on image
print(ocr_core(img))

## run the ocr on the pre-processed image
print(ocr_core(pp_img))
```
