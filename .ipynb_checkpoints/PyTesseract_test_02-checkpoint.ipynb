```python
import cv2
import numpy as np
import pytesseract
from matplotlib.pyplot import imshow
%matplotlib inline

config = "--psm 3"

def pre_processing(image):
    # Resize the image
    img = cv2.resize(image, None, fx=0.5, fy=0.5)

    # Convert image to grayscale
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Convert image to black and white (using adaptive threshold)
    img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 85, 11)
    
    return img

def ocr_core(image):    
    text = pytesseract.image_to_string(image, config=config, lang="eng")
    return text
```

```python
# Load and display the image
img = cv2.imread("book_page.jpg")
imshow(img)
```

```python
# run the ocr on image
print(ocr_core(img))
```

```python
# pre-process and displaay the image
pp_img = pre_processing(img)
imshow(pp_img)
```

```python
# run the ocr on the pre-processed image
print(ocr_core(pp_img))
```
